$P(A|B)=\frac {P(AB)}{P(B)}$, т.к. А и B независимые, то $P(AB)=P(A)$, т.е $P(A|B)=\frac {P(A)}{P(B)}$    
Вероятность, что на каждом кубике выпала 6: $P(A)=\frac {1}{6^3}=\frac {1}{216}$   
(a) по крайне мере на одной кости выпала "шестерка"   
Рассчиатем вероятность события, что хотя бы на одном кубике выпала 6, как 1 - вероятность невыпадения 6 ни на одном кубике  
$P(B)=1-\frac {5^3}{6^3}=\frac {91}{216}$   
Тогда $P(A|B)=\frac {1/216}{91/216}=\frac {1}{91}$   
(b) по крайне мере на двух костях выпало равное количество очков  
Рассчитаем вероятность события, что по крайне мере на двух костях выпало равное количество очков.    
Пусть на первой кости выпадет уникальное число, тогда на следюущей должно выпать любое отличное от первого (5/6) и на третьей кости такое же как на второй (1/6). Тогда $P(B)=\frac {1}{6}*\frac {5}{6}*\frac {6}{6}=\frac {30}{216} $   
Тогда $P(A|B)=\frac {1/216}{30/216}=\frac {1}{30}$

## 2. Распределения

```python
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

#### (А) Распределение Пуассона

```python
mu=10
mu1=12
mu2=14
x=np.arange(sts.poisson.ppf(0.01,mu1),sts.poisson.ppf(0.99,mu1))
f, (plt1,plt2)=plt.subplots(1,2)
f.set_figwidth(15)
plt1.plot(x,sts.poisson.pmf(x,mu),'bo')
plt1.plot(x,sts.poisson.pmf(x,mu1),'yo')
plt1.plot(x,sts.poisson.pmf(x,mu2),'ro')
plt1.set_title("Функция распредления")
plt1.set_xlabel('$mu$')
plt1.set_ylabel('$P(X=mu)$')
plt1.legend(["mu=10","mu=12","mu=14"])
plt2.plot(x,sts.poisson.cdf(x,mu),'bo')
plt2.plot(x,sts.poisson.cdf(x,mu1),'yo')
plt2.plot(x,sts.poisson.cdf(x,mu2),'ro')
plt2.set_xlabel('$\mu$')
plt2.set_ylabel('$P(X\leq mu)$')
plt2.set_title("Функция плотности")
plt2.legend(["mu=10","mu=12","mu=14"])
```

####  (B) Экспонинцеальное рапсределение 

```python
x=np.linspace(0,6,500)
f, (plt1,plt2)=plt.subplots(1,2)
f.set_figwidth(15)
plt1.plot(x,sts.expon.pdf(x,scale=0.5))
plt1.plot(x,sts.expon.pdf(x,scale=1))
plt1.plot(x,sts.expon.pdf(x,scale=2))
plt1.set_title("Функция распредления")
plt1.set_xlabel('$x$')
plt1.set_ylabel('$P(X)$')
plt1.legend(["$\lambda=2$","$\lambda=1$","$\lambda=0.5$"])
plt2.plot(x,sts.expon.cdf(x,scale=0.5))
plt2.plot(x,sts.expon.cdf(x,scale=1))
plt2.plot(x,sts.expon.cdf(x,scale=2))
plt2.set_title("Функция плотности")
plt2.set_xlabel('$x$')
plt2.set_ylabel('$P(X\leq x)$')
plt2.legend(["$\lambda=2$","$\lambda=1$","$\lambda=0.5$"])
```

#### (B) Логнормальное рапсределение 

```python
x=np.linspace(0,9,1000)
f, (plt1,plt2)=plt.subplots(1,2)
f.set_figwidth(15)
plt1.plot(x,sts.lognorm.pdf(x,s=0.25))
plt1.plot(x,sts.lognorm.pdf(x,s=0.5))
plt1.plot(x,sts.lognorm.pdf(x,s=1))
plt1.set_title("Функция распредления")
plt1.legend(["$\mu=0$ $shape=0.25$","$\mu=0$ $shape=0.5$","$\mu=0$ $shape=1$"])
plt1.set_xlabel('$x$')
plt1.set_ylabel('$P(X)$')

plt2.set_title("Функция плотности")
plt2.plot(x,sts.lognorm.cdf(x,s=0.25))
plt2.plot(x,sts.lognorm.cdf(x,s=0.5))
plt2.plot(x,sts.lognorm.cdf(x,s=1))
plt2.set_title("Функция плотности")
plt2.legend(["$\mu=0$ $shape=0.25$","$\mu=0$ $shape=0.5$","$\mu=0$ $shape=1$"])
plt2.set_xlabel('$x$')
plt2.set_ylabel('$P(X\leq x)$')
```
