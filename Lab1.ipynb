```python
import pandas as pnd
import scipy.stats as sts
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

## 1. Загрузить файл auto.csv

```python
auto_data=pnd.read_csv("auto.csv")
auto_data
```

## 2. Рассчитать основные статистики для переменных (среднее, медиана, мода, мин/макс, сред. отклонение)

```python
auto_data.describe()
```

```python
auto_data.mode()
```

## 3. Рассчитать и визуализировать корреляционную матрицу для переменных

```python
correllation_matrix=auto_data.corr()
```

```python
fig, ax = plt.subplots(figsize=(15,10))  
sns.heatmap(correllation_matrix, annot=True,ax=ax)
```

## 4.Визуализировать распределение переменной ‘Price’, рассчитать показатели skewness и kurtosis

```python
plt.hist(auto_data["price"])
plt.title("Histogram of Price")
plt.xlabel("Price")
print("Kurtosis {} Skewness {}".format(sts.kurtosis(auto_data["price"]),sts.skew(auto_data["price"])))
```

## 5. Визуализировать несколько взаимосвязей между переменными

```python
sns.pairplot(auto_data, vars=["price", "horsepower","engine-size"])
```

## 6. С использованием t-test статистики проверить взаимосвязи между несколькими переменными

```python
#scipy.stats.ttest_ind(auto_data["Price"],)

price_gas=auto_data.loc[auto_data["fuel-type"]=="gas","price"]
price_diesel=auto_data.loc[auto_data["fuel-type"]=="diesel","price"]
#print(np.array(price_gas).mean(), "  ",np.array(price_diesel).mean())
sts.ttest_ind(price_gas,price_diesel)
```
